```python
from scripts.datasetscripts import parse_concatenated_json, proportional_split, dump_data
import random

usual_dataset = parse_concatenated_json('/Users/artem.semidetnov/Desktop/MetaGenerator_v1/datasets/data_group_equation/data.json')

addon_dataset = parse_concatenated_json('/Users/artem.semidetnov/Desktop/MetaGenerator_v1/datasets/data1/addons.json')

datasets = proportional_split(usual_dataset, addon_dataset)
```

```python
total_train = datasets["train"]["A"][:10_000] + datasets["train"]["B"]
random.shuffle(total_train)
total_test = datasets["test"]["A"][:700] + datasets["test"]["B"]
random.shuffle(total_test)
total_val = datasets["val"]["A"][:700] + datasets["val"]["B"]
random.shuffle(total_val)
```
